```python
from pyspark.sql import SparkSession

# Path to the MySQL Connector/J JAR file
mysql_connector_jar = "C:/Users/HP/Downloads/mysql-connector-j-9.0.0/mysql-connector-j-9.0.0.jar"

# Initialize Spark session with the MySQL Connector/J
spark = SparkSession.builder \
    .appName("MariaDBIntegration") \
    .config("spark.jars", mysql_connector_jar) \
    .getOrCreate()

# JDBC URL for MariaDB
jdbc_url = "jdbc:mysql://127.0.0.1:3306/foodpricesdb"

# Table name in MariaDB
table_name = "foodprices"

# Database properties
db_properties = {
    "user": "root",
    "password": "root",  # Replace with your actual password
    "driver": "com.mysql.cj.jdbc.Driver"
}

# Load data from MariaDB
df_mariadb = spark.read.jdbc(jdbc_url, table_name, properties=db_properties)

# Show the schema and some sample data
df_mariadb.printSchema()
df_mariadb.show(5)
```
